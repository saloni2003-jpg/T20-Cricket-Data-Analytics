```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the total World Cups won by Team1
sns.barplot(data=df, x='Team1', y='Team1 Total WCs won')
plt.title('Total World Cups Won by Team1')
plt.xticks(rotation=90)
plt.xlabel('Team 1')
plt.ylabel('Total World Cups Won')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate the total World Cups won by both teams
team1_wc_won = df['Team1 Total WCs won'].sum()
team2_wc_won = df['Team2 Total WCs won'].sum()

# Prepare data for the pie chart
labels = ['Team1', 'Team2']
sizes = [team1_wc_won, team2_wc_won]
colors = ['#ff9999', '#66b3ff']

# Plotting the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, explode=(0.1, 0))
plt.title('Total World Cups Won by Team1 and Team2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate the total World Cups won by both teams
team1_wc_won = df['Team1 Total WCs won'].sum()
team2_wc_won = df['Team2 Total WCs won'].sum()

# Prepare data for the bar chart
teams = ['Team1', 'Team2']
wc_won = [team1_wc_won, team2_wc_won]

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=teams, y=wc_won, palette='viridis')
plt.title('Total World Cups Won by Team1 and Team2')
plt.xlabel('Teams')
plt.ylabel('Total World Cups Won')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate the total World Cups won by both teams
team1_wc_won = df['Team1 Total WCs won'].sum()
team2_wc_won = df['Team2 Total WCs won'].sum()

# Prepare data for the line graph
teams = ['Team1', 'Team2']
wc_won = [team1_wc_won, team2_wc_won]

# Plotting the line graph
plt.figure(figsize=(8, 6))
plt.plot(teams, wc_won, marker='o', linestyle='-', color='b', label='World Cups Won')
plt.title('Total World Cups Won by Team1 and Team2')
plt.xlabel('Teams')
plt.ylabel('Total World Cups Won')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate win percentage for Team1 and Team2
df['Team1 win % over Team2'] = (df['Team1 Total WCs won'] / df['Team1 Total WCs participated']) * 100
df['Team2 win % over Team1'] = (df['Team2 Total WCs won'] / df['Team2 Total WCs participated']) * 100

# Compare win percentages
team1_win_percent = df['Team1 win % over Team2'].mean()
team2_win_percent = df['Team2 win % over Team1'].mean()

print(f"Average Win Percentage for Team1: {team1_win_percent}%")
print(f"Average Win Percentage for Team2: {team2_win_percent}%")
```

```python
# Compare Batting and Bowling Rankings
avg_batting_ranking_team1 = df['Team1 Avg Batting Ranking'].mean()
avg_batting_ranking_team2 = df['Team2 Avg Batting Ranking'].mean()

avg_bowling_ranking_team1 = df['Team1 Avg Bowling Ranking'].mean()
avg_bowling_ranking_team2 = df['Team2 Avg Bowling Ranking'].mean()

print(f"Average Batting Ranking for Team1: {avg_batting_ranking_team1}")
print(f"Average Batting Ranking for Team2: {avg_batting_ranking_team2}")
print(f"Average Bowling Ranking for Team1: {avg_bowling_ranking_team1}")
print(f"Average Bowling Ranking for Team2: {avg_bowling_ranking_team2}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='Team1 Avg Batting Ranking',
    y='Team1 Avg Bowling Ranking',
    hue='Winner',
    palette='viridis',
    s=80
)
plt.title("Team1 Batting vs Bowling Ranking")
plt.xlabel("Avg Batting Ranking")
plt.ylabel("Avg Bowling Ranking")
plt.grid(True)
plt.show()
```

```python
win_counts = df['Winner'].value_counts()
print("Top Winning Teams:\n", win_counts.head())
```

```python
df['Match Date'] = pd.to_datetime(df['Match Date'])
df['Year'] = df['Match Date'].dt.year

matches_per_year = df.groupby('Year').size()
matches_per_year.plot(kind='bar', title='Matches per Year', figsize=(10,5))
```

```python
upsets = df[(df['Team1 Avg Batting Ranking'] > df['Team2 Avg Batting Ranking']) & (df['Winner'] == df['Team1'])]
print("Upsets where Team1 had worse batting ranking but still won:", upsets.shape[0])
```
